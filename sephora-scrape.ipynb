```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

base_url = "https://www.ulta.com/shop/fragrance/womens-fragrance/perfume?page={}"


data = []

def extract_product(card):
    # Perfume link (if present)
    link_tag = card.find("a", class_="pal-c-Link")
    link = link_tag.get("href") if link_tag else None
    if link and link.startswith("/"):
        link = "https://www.ulta.com" + link

    # Image (check src, data-src, srcset)
    img_tag = card.find("img")
    img_src = None
    if img_tag:
        img_src = img_tag.get("src")
        if (img_src and img_src.startswith("data:")) or not img_src:
            img_src = img_tag.get("data-src") or img_tag.get("srcset")

    # Brand
    brand_tag = card.find("span", class_="ProductCard__brand")
    brand = brand_tag.get_text(strip=True) if brand_tag else None

    # Product name
    product_tag = card.find("span", class_="ProductCard__product")
    product_name = product_tag.get_text(strip=True) if product_tag else None

    # Price
    price_tag = card.find("div", class_="ProductPricing")
    price = price_tag.get_text(strip=True) if price_tag else None

    if brand or product_name:  # ensure it's valid
        return {
            "Brand": brand,
            "Product": product_name,
            "Price": price,
            "Image_Link": img_src,
            "Product_Link": link
        }
    return None


# --- Loop through pages 1–11 ---
for page in range(1, 12):
    url = base_url.format(page)
    print(f"Scraping {url} ...")

    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    # Case 1: li-based product cards
    cards1 = soup.find_all("li", class_="ProductListingResults__productCard")
    for c in cards1:
        prod = extract_product(c)
        if prod:
            data.append(prod)

    # Case 2: div-based product cards
    cards2 = soup.find_all("div", class_="ProductCard__content")
    for c in cards2:
        prod = extract_product(c.parent)
        if prod:
            data.append(prod)

    time.sleep(1)  # be polite, avoid hammering server


# --- Build DataFrame ---
df = pd.DataFrame(data).drop_duplicates(subset=["Brand", "Product"]).reset_index(drop=True)

print(df.head())
print("Total unique perfumes scraped:", len(df))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"
}

def get_summary_details(product_url):
    """Fetch product page and extract summary + details text."""
    try:
        r = requests.get(product_url, headers=headers, timeout=10)
        soup = BeautifulSoup(r.text, "html.parser")
    except Exception as e:
        print("Error fetching:", product_url, e)
        return None

    summary = None
    details = {}

    # Summary: find h2 “Summary” then the following <p>
    h2 = soup.find("h2", string="Summary")
    if h2:
        p = h2.find_next_sibling("p")
        if p:
            summary = p.get_text(strip=True)

    # Details: find the “Details” section if exists
    # Use the headers like “Fragrance Family”, “Key Notes”, “Benefits”, “How To Use”, “Ingredients”
    # They appear in markdown/HTML headings like <h4> or <h3> following
    # We'll search for those headings and their following lists/text.

    # Example mapping of detail names to tag strings
    detail_keys = ["Fragrance Family", "Key Notes", "Benefits", "How To Use", "Ingredients"]
    for key in detail_keys:
        # find an element whose text matches key (or starts with it)
        # sometimes “Fragrance Family” is under <h4> or <h3>
        hdr = soup.find(lambda tag: tag.name in ["h3", "h4", "h5"] and key in tag.get_text())
        if hdr:
            # Get next sibling(s) until next header or break
            texts = []
            for sib in hdr.next_siblings:
                # Stop when we hit another header (h3/h4 etc)
                if getattr(sib, "name", None) in ["h3", "h4", "h5"]:
                    break
                # If it's a <ul> or <p> or <div> with text, gather it
                if hasattr(sib, "get_text"):
                    txt = sib.get_text(strip=True)
                    if txt:
                        texts.append(txt)
            details[key] = " ".join(texts) if texts else None
        else:
            details[key] = None

    return {
        "summary": summary,
        **details
    }


# Example of integrating into your main scraping loop
# Suppose you already built df with columns including “Product_Link”:

# After building df (Brand, Product, Price, Image_Link, Product_Link):
summaries = []
for idx, row in df.iterrows():
    link = row["Product_Link"]
    if link:
        info = get_summary_details(link)
    else:
        info = None
    summaries.append(info)
    time.sleep(0.5)  # be polite

# Turn the list of dicts into a DataFrame
df_details = pd.DataFrame(summaries)
# Concatenate to main df
df_full = pd.concat([df.reset_index(drop=True), df_details], axis=1)

# Then df_full will have columns:
# Brand, Product, Price, Image_Link, Product_Link,
# summary, Fragrance Family, Key Notes, Benefits, How To Use, Ingredients

print(df_full.head())
```

```python
def get_extra_details(product_url):
    try:
        r = requests.get(product_url, headers=headers, timeout=10)
        soup = BeautifulSoup(r.text, "html.parser")
    except Exception as e:
        print("Error fetching:", product_url, e)
        return {"Benefits": None, "How To Use": None, "Ingredients": None}

    results = {"Benefits": None, "How To Use": None, "Ingredients": None}

    for key in results.keys():
        hdr = soup.find(lambda tag: tag.name in ["h2", "h3", "h4"]
                        and tag.get_text(strip=True).startswith(key))
        if hdr:
            # Look specifically for the next Markdown div
            markdown_div = hdr.find_next_sibling("div", class_="Markdown")
            if markdown_div:
                results[key] = markdown_div.get_text(" ", strip=True)

    return results
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_perfume_details(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/124.0.0.0 Safari/537.36"
    }
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser')

    # Initialize
    benefits = how_to_use = ingredients = None

    # --- BENEFITS ---
    benefits_header = soup.find('h4', string=lambda s: s and 'Benefits' in s)
    if benefits_header:
        ul = benefits_header.find_next('ul')
        if ul:
            benefits = [li.get_text(strip=True) for li in ul.find_all('li')]
            benefits = " | ".join(benefits)

    # --- HOW TO USE ---
    how_to_use_header = soup.find('h3', string=lambda s: s and 'How To Use' in s)
    if how_to_use_header:
        section = how_to_use_header.find_parent('div', class_='pal-c-Accordion__header')
        if section:
            body = section.find_next('section', class_='pal-c-Accordion__body')
            text_block = body.find('div', {'data-test': 'markdown'}) if body else None
            if text_block:
                how_to_use = text_block.get_text(separator=' ', strip=True)

    # --- INGREDIENTS ---
    ingredients_header = soup.find('h3', string=lambda s: s and 'Ingredients' in s)
    if ingredients_header:
        section = ingredients_header.find_parent('div', class_='pal-c-Accordion__header')
        if section:
            body = section.find_next('section', class_='pal-c-Accordion__body')
            text_block = body.find('div', {'data-test': 'markdown'}) if body else None
            if text_block:
                ingredients = text_block.get_text(separator=' ', strip=True)

    return {
        "ProductLink": url,
        "Benefits": benefits,
        "HowToUse": how_to_use,
        "Ingredients": ingredients
    }

# Example usage:
test_url = "https://www.ulta.com/p/flora-gorgeous-gardenia-eau-de-parfum-intense-pimprod2052375?sku=2640483"
details = scrape_perfume_details(test_url)
print(details)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_perfume_details_fixed(url):
    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/124.0.0.0 Safari/537.36"
        )
    }
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    benefits = how_to_use = ingredients = None

    # --- BENEFITS ---
    benefits_header = soup.find("h4", string=lambda s: s and "Benefits" in s)
    if benefits_header:
        ul = benefits_header.find_next("ul")
        if ul:
            benefits = " | ".join([li.get_text(strip=True) for li in ul.find_all("li")])

    # --- HOW TO USE ---
    how_to_use_header = soup.find("h3", string=lambda s: s and "How To Use" in s)
    if how_to_use_header:
        header_div = how_to_use_header.find_parent("div", class_="pal-c-Accordion__header")
        if header_div:
            # Correct: find the NEXT SIBLING <section>, not "find_next"
            body_section = header_div.find_next_sibling("section", class_="pal-c-Accordion__body")
            if body_section:
                markdown_div = body_section.find("div", {"data-test": "markdown"})
                if markdown_div:
                    how_to_use = markdown_div.get_text(separator=" ", strip=True)

    # --- INGREDIENTS ---
    ingredients_header = soup.find("h3", string=lambda s: s and "Ingredients" in s)
    if ingredients_header:
        header_div = ingredients_header.find_parent("div", class_="pal-c-Accordion__header")
        if header_div:
            # Again, correct sibling path
            body_section = header_div.find_next_sibling("section", class_="pal-c-Accordion__body")
            if body_section:
                markdown_div = body_section.find("div", {"data-test": "markdown"})
                if markdown_div:
                    ingredients = markdown_div.get_text(separator=" ", strip=True)

    return {
        "ProductLink": url,
        "Benefits": benefits,
        "HowToUse": how_to_use,
        "Ingredients": ingredients,
    }


# Example test
url = "https://www.ulta.com/p/flora-gorgeous-gardenia-eau-de-parfum-intense-pimprod2052375?sku=2640483"
print(scrape_perfume_details_fixed(url))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm  # progress bar

def scrape_perfume_details_fixed(url):
    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/124.0.0.0 Safari/537.36"
        )
    }
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    benefits = how_to_use = ingredients = None

    # --- BENEFITS ---
    benefits_header = soup.find("h4", string=lambda s: s and "Benefits" in s)
    if benefits_header:
        ul = benefits_header.find_next("ul")
        if ul:
            benefits = " | ".join([li.get_text(strip=True) for li in ul.find_all("li")])

    # --- HOW TO USE ---
    how_to_use_header = soup.find("h3", string=lambda s: s and "How To Use" in s)
    if how_to_use_header:
        header_div = how_to_use_header.find_parent("div", class_="pal-c-Accordion__header")
        if header_div:
            body_section = header_div.find_next_sibling("section", class_="pal-c-Accordion__body")
            if body_section:
                markdown_div = body_section.find("div", {"data-test": "markdown"})
                if markdown_div:
                    how_to_use = markdown_div.get_text(separator=" ", strip=True)

    # --- INGREDIENTS (robust fix) ---
    # Find all accordion headers, then locate the one labeled "Ingredients"
    accordions = soup.find_all("div", class_="pal-c-Accordion__header")
    for header in accordions:
        title_el = header.find("h3")
        if title_el and "ingredients" in title_el.get_text(strip=True).lower():
            section = header.find_next_sibling("section", class_="pal-c-Accordion__body")
            if section:
                markdown_div = section.find("div", {"data-test": "markdown"})
                if markdown_div:
                    ingredients = markdown_div.get_text(separator=" ", strip=True)
                else:
                    # fallback: get raw HTML string if text extraction fails
                    ingredients = str(section)
            break

    return {
        "ProductLink": url,
        "Benefits": benefits,
        "HowToUse": how_to_use,
        "Ingredients": ingredients,
    }


# === Apply to all product links in your DataFrame ===
# Assume df_details has a column "Product_Link"
results = []

for link in tqdm(df_details["Product_Link"], desc="Scraping perfume details"):
    try:
        details = scrape_perfume_details_fixed(link)
        results.append(details)
    except Exception as e:
        results.append({"ProductLink": link, "Benefits": None, "HowToUse": None, "Ingredients": None})
        print(f"Error scraping {link}: {e}")

# Convert to DataFrame and merge
df_new_details = pd.DataFrame(results)

# Merge new columns back into your original df_details
df_details = df_details.merge(df_new_details, left_on="Product_Link", right_on="ProductLink", how="left")

# Drop redundant ProductLink column if desired
#df_details.drop(columns=["ProductLink"], inplace=True)

df_details.head()
```
